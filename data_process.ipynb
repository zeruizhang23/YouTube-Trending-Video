```python
import pandas as pd
import json
```

```python
us = pd.read_csv('../archive/US_youtube_trending_data.csv')
us['country'] = 'US'


with open('../archive/US_category_id.json', 'r') as file:
    categories_json = json.load(file)

us_category_mapping = {int(item['id']): item['snippet']['title'] for item in categories_json['items']}
us['categoryId'] = us['categoryId'].astype(int)
us['categoryName'] = us['categoryId'].map(us_category_mapping)

us.head()
```

```python
canada = pd.read_csv('../archive/CA_youtube_trending_data.csv')
canada['country'] = 'Canada'

with open('../archive/CA_category_id.json', 'r') as file:
    categories_json = json.load(file)

ca_category_mapping = {int(item['id']): item['snippet']['title'] for item in categories_json['items']}
canada['categoryId'] = canada['categoryId'].astype(int)
canada['categoryName'] = canada['categoryId'].map(ca_category_mapping)

canada.head()
```

```python
mexico = pd.read_csv('../archive/MX_youtube_trending_data.csv')
mexico['country'] = 'Mexico'

with open('../archive/MX_category_id.json', 'r') as file:
    categories_json = json.load(file)

mx_category_mapping = {int(item['id']): item['snippet']['title'] for item in categories_json['items']}
mexico['categoryId'] = mexico['categoryId'].astype(int)
mexico['categoryName'] = mexico['categoryId'].map(mx_category_mapping)

mexico.head()
```

```python
df = pd.concat([us, canada, mexico])
df.head()
df.tail()
```

```python

df['trending_date'] = pd.to_datetime(df['trending_date'], format='%Y-%m-%dT%H:%M:%SZ')
df['publishedAt'] = pd.to_datetime(df['publishedAt'], format='%Y-%m-%dT%H:%M:%SZ')

# current_date = pd.to_datetime('today').date()
# recent_data = combined_data[combined_data['trending_date'] >= current_date - pd.Timedelta(days=90)]
# recent_data.tail()

most_recent_date = df['trending_date'].max()
three_months_ago = most_recent_date - pd.DateOffset(months=3)
filtered_df = df[df['trending_date'] >= three_months_ago]

```

```python
#output file
filtered_df.to_csv('YouTube-Trending-Video.csv', index=False)
```
